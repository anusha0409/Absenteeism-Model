```python
import pandas as pd
import numpy as np
```

```python
datap=pd.read_csv('Absent_preprocessed.csv')
```

```python
datap.head()
```

```python
datap=datap.drop('Date',axis=1)
```

```python
datap.columns.values
```

```python
datap.columns
```

```python
array=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month value', 'Day of the week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
array
```

```python
datap=datap[array]
```

```python
datap
```

```python
datap['Absenteeism Time in Hours'].median()
```

```python
targets=np.where(datap['Absenteeism Time in Hours']>3,1,0)
```

```python
targets
```

```python
datap['excess']=targets
```

```python
datap.head()
```

```python
targets.sum()/targets.shape[0]
```

```python
datatargets=datap.drop(['Absenteeism Time in Hours','Day of the week','Distance to Work',"Daily Work Load Average"],axis=1)
```

```python
datatargets.head()
```

```python
datatargets.shape
```

```python
datatargets.iloc[:,:-1]
```

```python
unscaled=datatargets.iloc[:,:-1]
```

```python
unscaled
```

STANDARDIZE


```python
from sklearn.preprocessing import StandardScaler
absent_scaler=StandardScaler()
```

```python
absent_scaler.fit(unscaled)
```

```python
scaled=absent_scaler.transform(unscaled)
```

```python
scaled
```

```python
scaled.shape
```

SPLIT THE DATA INTO TEST AND TRAIN AND SHUFFLE

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled,targets)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled,targets,train_size=0.8,random_state=20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

MODELLING


```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

MANUAL ACCURACY

```python
model_outputs=reg.predict(x_train)
```

```python
model_outputs
```

```python
targets
```

```python
model_outputs==y_train
```

```python
unscaled.columns.values
```

```python
feature=unscaled.columns.values
```

```python
summary=pd.DataFrame(columns=['feature name'],data=feature)
summary['Coefficient']=np.transpose(reg.coef_)
```

```python
summary
```

```python
summary.index=summary.index+1
summary.loc[0]=['Intercept',reg.intercept_[0]]
summary=summary.sort_index()
summary
```

```python
summary["odds ratio"]=np.exp(summary.Coefficient)
```

```python
summary
```

```python
summary.sort_values('odds ratio',ascending =False)
```

```python
reg.score(x_test,y_test)
```
